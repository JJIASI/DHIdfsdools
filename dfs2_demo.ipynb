```python
from dfs2 import dfs2
```

`.read_dfs2(step=0)` step value default is `0`.  
You can change step value, but step value must be `int`.  
Output parameters `x`, `y` format is `NumPy array`   
`z` is dictionary

```python
path = 'dfs2 FilePath'
dfs2_obj = dfs2(path)
x,y,z = dfs2_obj.read_dfs2()
z['Bathymetry']['2002-12-31 23:59:59']
```

This Example data is from MIKE_21 example file(Torsemide_Harbour).  
`x`, `y` must do `numpy.meshgrid`,because `x`, `y` is one dimension.

```python
import numpy as np
import matplotlib.pyplot as plt

x, y = np.meshgrid(x,y)
z1 = z['Bathymetry']['2002-12-31 23:59:59']
plt.figure()
plt.contourf(x, y, z1,100)
plt.colorbar()
plt.axis('scaled')
plt.show()
```
